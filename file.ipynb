```python
df['Session Duration'] = (df['Logout Time'] - df['Login Time']).dt.total_seconds()/60
```

```python
df['Logout hour'] = df['Logout Time'].dt.hour
```

```python
def create_target_label(target_hour):
    df['Engaged'] = df['Login hour'].apply(lambda x:1 if x == target_hour else 0 )

    create_target_label(target_hour=15)



```

```python
df = df[['NPI', 'State', 'Login Time', 'Logout Time', 'Usage Time (mins)',
       'Region', 'Speciality', 'Count of Survey Attempts', 'Login hour',
       'Session Duration', 'Logout hour', 'Engaged']]
```

```python
df.columns 
```

```python
df['Login hour'].dtype
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['State'] = encoder.fit_transform(df['State'])
df['Region'] = encoder.fit_transform(df['Region'])
df['Speciality'] = encoder.fit_transform(df['Speciality'])
```

```python
X = df[['State', 'Region', 'Speciality', 'Login hour', 'Logout hour', 'Session Duration', 'Usage Time (mins)', 'Count of Survey Attempts']]
Y = df['Engaged']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train , Y_test = train_test_split(X, Y , test_size= 0.2, random_state= 2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=2)
model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_pred, Y_test)
accuracy
```

```python
confusion = confusion_matrix(y_pred,Y_test)
confusion
```

```python


classification = classification_report(Y_test, y_pred)
print(classification)
```

```python
train_y_pred = model.predict(X_train)
train_accuracy = accuracy_score(Y_train, train_y_pred)
train_accuracy
```

```python
train_confusion = confusion_matrix(Y_train , train_y_pred)
train_confusion
```

```python
train_classification = classification_report(Y_train, train_y_pred)
print(train_classification)
```

```python
sns.heatmap(confusion, annot= True, fmt="d",cmap="Blues",xticklabels=["Not Engaged", "Engaged"], yticklabels= ["Not Engaged", "Engaged"])
plt.title("Confusion matrix")
plt.show()
```

```python
sns.heatmap(train_confusion, annot=True,fmt="d",cmap="Blues",xticklabels=["Not Engaged","Engaged"], yticklabels=["Not Engaged", "Engaged"])
plt.title("Confusion Matrix")
plt.show()
```

```python
important = model.feature_importances_
features = X.columns
indicies = important.argsort()[::-1]

plt.figure(figsize=(8,5))
plt.title("Feature Importance")
plt.barh(range(X.shape[1]), important[indicies], align="center")
plt.yticks(range(X.shape[1]), [features[i] for i in indicies])
plt.show()
```
